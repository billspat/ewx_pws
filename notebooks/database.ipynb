```python
def dict_factory(cursor, row):
    d = {}
    for idx, col in enumerate(cursor.description):
        d[col[0]] = row[idx]
    return d

def list_tables(conn):
    # cursor = conn.cursor()
    result = conn.execute("PRAGMA table_list;")
    table_list =  result.fetchall()
    result.close()
    return(table_list)

def table_columns(conn, tablename):
    # cursor = conn.cursor()
    columnsQuery = ("PRAGMA table_info(%s)" % (tablename))
    result = conn.execute(columnsQuery)
    columns = [column for column in result.fetchall()]
    result.close()
    return(columns)

print("tables")
print(list_tables(weatherstore.connection))
print("columns")
print(table_columns(weatherstore.connection, 'STATIONS'))
```

```python
from sqlalchemy import engine,create_engine
from os import getenv
from dotenv import load_dotenv
load_dotenv('../.env')

PWSDBURI=getenv('PWSDBURI')
print(PWSDBURI)
if not PWSDBURI:
    raise ValueError("PWSDBURI is not set")
    
engine = create_engine(PWSDBURI)
```

```python
from sqlalchemy import MetaData, inspect


# method A
inspector = inspect(engine)
for table_name in inspector.get_table_names():
   print(table_name)
   print('-----')
   for column in inspector.get_columns(table_name):
       print("Column: %s" % column['name'])

# method B, for postgresql
schema="public"
with engine.connect() as conn:
        meta = MetaData(conn, schema=schema)
        meta.reflect(views=True)
        tables = (list(meta.tables.keys()))

print(tables)
```
